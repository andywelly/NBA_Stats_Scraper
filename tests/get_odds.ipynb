```python
import pandas as pd
import requests
pd.set_option('display.max_columns', None)
import time
import numpy as np
```

```python
odds_url = 'https://cdn.nba.com/static/json/liveData/odds/odds_todaysGames.json'
```

```python
odds_data = requests.get(url = odds_url).json()
```

```python
odds_data
```

```python
odds_list = odds_data['games']

for game in odds_list:
    gameId = game['gameId']
    markets = game['markets'][0]
    books = markets['books'][0]['outcomes']
    print(books)
```
